# Load the data

Place the files in the ../../05_src/data/fires/ directory and load the appropriate file. 

```python
# Load the environment variables and libraries 
%load_ext dotenv
%dotenv 

import os
import pandas as  pd 
from glob import glob
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PowerTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
import pickle
import shap 
```

```python
# Load data
columns = [
    'coord_x', 'coord_y', 'month', 'day', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain', 'area' 
]
fires_dt = (pd.read_csv('../../05_src/data/fires/forestfires.csv', header = 0, names = columns))
fires_dt.info()
```

# Get X and Y

Create the features data frame and target data.

```python
# Create the target dataframe 
y = pd.DataFrame(fires_dt["area"])
y
```

```python
# Create the features dataframe
x=  fires_dt.drop(columns=["area"])
x
```

# Preprocessing

Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

- Numerical variables

    * (Preproc 1 and 2) Scaling: use a scaling method of your choice (Standard, Robust, Min-Max). 
    * Preproc 2 only: 
        
        + Choose a transformation for any of your input variables (or several of them). Evaluate if this transformation is convenient.
        + The choice of scaler is up to you.

- Categorical variables: 
    
    * (Preproc 1 and 2) Apply [one-hot encoding](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) where appropriate.


+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python
# Define column names
num_col_names = ['coord_x', 'coord_y', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain']
cat_col_names = ['month', 'day']

# Extract the columns want to apply multiple transformations 
dmc_col = fires_dt['dmc']
dmc_col_names = ['dmc']
```

```python

# Transform the numerical variables and categorical variables for preproc1
preproc1 = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_col_names),  # Numeric transformation
        ('cat', OneHotEncoder(), cat_col_names)   # Categorical transformation
    ]
)
preproc1
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
# Create a pipline and apply multiple transformations to certain columns
preproc_mt = Pipeline([
    ('scale', StandardScaler()),
    ('power', PowerTransformer())
])


# Transform the numerical variables and categorical variables for preproc1
preproc2 = ColumnTransformer([
    ('num', StandardScaler(), num_col_names),
    ('cat', OneHotEncoder(), cat_col_names),
    ('dmc', preproc_mt, dmc_col_names) 
])
preproc2
```

## Model Pipeline


Create a [model pipeline](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html): 

+ Add a step labelled `preprocessing` and assign the Column Transformer from the previous section.
+ Add a step labelled `regressor` and assign a regression model to it. 

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Choose a more advance regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Both model choices are up to you, feel free to experiment.

```python
# Pipeline A = preproc1 + baseline
model_pipeline_A = Pipeline(
    steps=[
        ('preprocessing', preproc1),  # Preprocessing step
        ('regressor', LinearRegression())  # Regression model
    ]
)
```

```python
# Pipeline B = preproc2 + baseline
model_pipeline_B = Pipeline(
    steps=[
        ('preprocessing', preproc2),  # Preprocessing step
        ('regressor', LinearRegression())  # Regression model
    ]
)
```

```python
# Pipeline C = preproc1 + advanced model
model_pipeline_C = Pipeline(
    steps=[
        ('preprocessing', preproc1),  # Preprocessing step
        ('regressor', GradientBoostingRegressor())  # Regression model
    ]
)

```

```python
# Pipeline D = preproc2 + advanced model
model_pipeline_D = Pipeline(
    steps=[
        ('preprocessing', preproc2),  # Preprocessing step
        ('regressor', GradientBoostingRegressor())  # Regression model
    ]
)
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.
+ Experiment with at least four value combinations per pipeline.

```python
# Define the paramter grid for linear regression model pipeline A 
param_grid_A = {
    'regressor__fit_intercept': [True, False]  
    }

# Define the parameter grid for linear regression model pipeline B
param_grid_B = {
    'regressor__fit_intercept': [True, False], 
    'regressor__positive': [True, False]  
    }

# Define the parameter grid for advanced regressor model pipeline C 
param_grid_C = {
    'regressor__n_estimators': [100, 200],
    'regressor__learning_rate': [0.01, 0.1],
    'regressor__max_depth': [3, 5],
    'regressor__min_samples_split': [2, 5],
    'regressor__subsample': [0.8, 1.0]
}

# Define the parameter grid for advanced regressor model pipeline D 
param_grid_D = {
    'regressor__n_estimators': [200, 300],
    'regressor__learning_rate': [0.01, 0.1],
    'regressor__max_depth': [3, 6],
    'regressor__min_samples_split': [2, 6],
    'regressor__subsample': [0.8, 1.0]
}

# Set up the scoring metrics 
scoring = ['neg_mean_squared_error', 'neg_root_mean_squared_error', 'r2', 'neg_mean_absolute_error']

#Set up the splitting strategy 
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.25, random_state = 123)
```

```python
# Find the best combination of parameters 
grid_cv_A = GridSearchCV(
    estimator=model_pipeline_A,
    param_grid=param_grid_A,
    scoring=scoring,
    cv=5,
    refit="neg_root_mean_squared_error" # Metric to select the best model
)

results_A = grid_cv_A.fit(X_train, Y_train)
```

```python
# Find the best combination of parameters 
grid_cv_B = GridSearchCV(
    estimator=model_pipeline_B,
    param_grid=param_grid_B,
    scoring=scoring,
    cv=5,
    refit="neg_root_mean_squared_error" # Metric to select the best model
)


results_B = grid_cv_B.fit(X_train, Y_train)
```

```python
# Find the best combination of parameters 
grid_cv_C = GridSearchCV(
    estimator=model_pipeline_C,
    param_grid=param_grid_C,
    scoring=scoring,
    cv=5,
    refit="neg_root_mean_squared_error"
)


results_C = grid_cv_C.fit(X_train, Y_train)
```

```python
# Find the best combination of parameters 
grid_cv_D = GridSearchCV(
    estimator=model_pipeline_D,
    param_grid=param_grid_D,
    scoring=scoring,
    cv=5,
    refit = "neg_root_mean_squared_error"
)

results_D = grid_cv_D.fit(X_train, Y_train)
```

# Evaluate

+ Which model has the best performance?

```python
# Obtain the best models 
best_model_A = results_A.best_estimator_
best_model_B = results_B.best_estimator_
best_model_C = results_C.best_estimator_
best_model_D = results_D.best_estimator_

# Obtain the scores from the best models and convert the values back to positive 
performance_A = -results_A.best_score_
performance_B = -results_B.best_score_
performance_C = -results_C.best_score_
performance_D = -results_D.best_score_

# Find the best score out of all the models 
model_performances = {
    'Model A': performance_A,
    'Model B': performance_B,
    'Model C': performance_C,
    'Model D': performance_D
}

best_model_name = min(model_performances)
print(best_model_name)
```

# Export

+ Save the best performing model to a pickle file.

```python
# Rename the best model 
best_model = best_model_A
```

```python
# Save the model as a pickle file 
with open('best_model.pkl', 'wb') as file:
    pickle.dump(best_model, file)
```
